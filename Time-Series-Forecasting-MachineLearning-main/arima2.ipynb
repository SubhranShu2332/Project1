```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load the past data into a Pandas DataFrame
df = pd.read_csv('final_sd.csv', parse_dates=['Date'], index_col='Date')

# Plot the past data to visualize any trends or patterns
plt.plot(df)
plt.title('Past Laid Off Data')
plt.xlabel('Time')
plt.ylabel('Laid Off Employees')
plt.show()

# Fit an ARIMA model to the past data to make predictions for the future
model = ARIMA(df, order=(0, 1, 1))  # (p, d, q)
results = model.fit()

# Make predictions for the future using the fitted model
future_dates = pd.date_range(start='2023-03-01', end='2025-04-01')
future_predictions = results.predict(start=len(df), end=len(df)+len(future_dates)-1, dynamic=False)

# Plot the past data and the predicted values for the future
plt.plot(df)
plt.plot(future_dates, future_predictions, color='red')
plt.title('Past Laid Off Data and Future Predictions')
plt.xlabel('Time')
plt.ylabel('Laid Off Employees')
plt.show()
```
