```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset from a CSV file
df = pd.read_csv('LayOff India-Web-Development.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

df['industry_grouped'] = df.groupby('industry')['total_laid_off'].transform('sum')

import matplotlib.pyplot as plt

# Group the data by year and industry and sum the total laid off
df_grouped = df.groupby([pd.Grouper(key='date', freq='Y'), 'industry'])['total_laid_off'].sum().reset_index()

# Pivot the dataframe to have industry as columns and year as rows
df_pivot = df_grouped.pivot(index='date', columns='industry', values='total_laid_off')

# Plot the data
df_pivot.plot.line(figsize=(20, 10))

# Set the title and axis labels
plt.title('Total Laid Off by Industry')
plt.xlabel('Year')
plt.ylabel('Total Laid Off')
plt.show()


```
