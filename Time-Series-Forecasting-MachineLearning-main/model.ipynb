```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the past data into a Pandas DataFrame
df = pd.read_csv('webDev5.csv', parse_dates=['Date'], index_col='Date')

# Compute the simple moving average of the past data using a window size of 7 (one week)
sma = df.rolling(window=7).mean()

# Plot the past data and the simple moving average
plt.plot(df)
plt.plot(sma, color='red')
plt.title('Past Laid Off Data and Simple Moving Average')
plt.xlabel('Time')
plt.ylabel('Laid Off Employees')
plt.show()

# Use the simple moving average to make predictions for the future
last_value = df.iloc[-1]['Layoffs']
future_predictions = [sma.iloc[-1]['Layoffs']] * 367  # Predict the next 30 days using the last moving average value

# Plot the past data, the simple moving average, and the predicted values for the future
plt.plot(df)
#plt.plot(sma, color='transperant')
plt.plot(pd.date_range(start='2023-03-01', end='2024-03-01'), future_predictions, color='green')
plt.title('Past Laid Off Data, Simple Moving Average, and Future Predictions')
plt.xlabel('Time')
plt.ylabel('Laid Off Employees')
plt.show()
```
